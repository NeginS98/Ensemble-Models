## **Family Size and Ticket Type**

Refrence for def that I wrote below :
https://www.kaggle.com/code/volhaleusha/titanic-tutorial-encoding-feature-eng-81-8/notebook

```python
# Create function that take name and separates it into title, family name and deletes all puntuation from name column:
def name_sep(data):
    families=[]
    titles = []
    new_name = []
    #for each row in dataset:
    for i in range(len(data)):
        name = data.iloc[i]
        # extract name inside brakets into name_bracket:
        if '(' in name:
            name_no_bracket = name.split('(')[0] 
        else:
            name_no_bracket = name
            
        family = name_no_bracket.split(",")[0]
        title = name_no_bracket.split(",")[1].strip().split(" ")[0]
        
        #remove punctuations accept brackets:
        import string
        for c in string.punctuation:
            name = name.replace(c,"").strip()
            family = family.replace(c,"").strip()
            title = title.replace(c,"").strip()
            
        families.append(family)
        titles.append(title)
        new_name.append(name)
            
    return families, titles, new_name  
```

```python
def ticket_sep(data_ticket):
    ticket_type = []

    for i in range(len(data_ticket)):

            ticket =data_ticket.iloc[i]
            import string
            for c in string.punctuation:
                ticket = ticket.replace(c,"")
                splited_ticket = ticket.split(" ")   
            if len(splited_ticket) == 1:
                ticket_type.append('NO')
            else: 
                ticket_type.append(splited_ticket[0])
    return ticket_type 
```

### Define Class For modeling by various algorithms

Refrence : https://towardsdatascience.com/ensemble-models-5a62d4f4cb0c

```python
# Model Class to be used for different ML algorithms
class ClassifierModel(object):
    def __init__(self, clf, params=None):
        self.clf = clf(**params)
    def fit(self,x,y):
        return self.clf.fit(x,y)

    def feature_importances(self,x,y):
        return self.clf.fit(x,y).feature_importances_

    def predict(self, x):
        return self.clf.predict(x)
    def train(self, x_train, y_train):
        self.clf.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
def trainModel(model, x_train, y_train, x_test, n_folds, seed):
    cv = KFold(n_splits= n_folds, random_state=seed)
    scores = cross_val_score(model.clf, x_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.25)
```

### Random Forest Classifier

```python
# Random Forest parameters
from sklearn.ensemble import RandomForestClassifier
rf_params = {
    'n_estimators': 400,
    'max_depth': 5,
    'min_samples_leaf': 3,
    'max_features' : 'sqrt',
}
rfc_model = ClassifierModel(clf=RandomForestClassifier, params=rf_params)
rfc_scores = trainModel(rfc_model,X_train, y_train,X_test, 5, None)
rfc_scores
```

### Extra Trees Classifier

```python
# Extra Trees Parameters
from sklearn.ensemble import ExtraTreesClassifier
et_params = {
    'n_jobs': -1,
    'n_estimators':400,
    'max_depth': 5,
    'min_samples_leaf': 2,
}
etc_model = ClassifierModel(clf=ExtraTreesClassifier, params=et_params)
etc_scores = trainModel(etc_model,X_train, y_train, X_test, 5, None)
etc_scores
```

### **AdaBoost Classifier**

```python
# AdaBoost parameters
from sklearn.ensemble import AdaBoostClassifier
ada_params = {
    'n_estimators': 400,
    'learning_rate' : 0.65
}
ada_model = ClassifierModel(clf=AdaBoostClassifier, params=ada_params)
ada_scores = trainModel(ada_model,X_train, y_train, X_test, 5, None)
ada_scores
```

### XGBoost Classifier

```python
# Gradient Boosting parameters
from sklearn.ensemble import GradientBoostingClassifier
gb_params = {
    'n_estimators': 400,
    'max_depth': 6,
}
gbc_model = ClassifierModel(clf=GradientBoostingClassifier, params=gb_params)
gbc_scores = trainModel(gbc_model,X_train, y_train, X_test, 5, None) 
gbc_scores
```

### Getting features importance 

```python
lst_mdl = [rfc_model, etc_model, ada_model, gbc_model]
lst_mdl_name = ['Random Forest Score','Extra Trees score','AdaBoost score','Gradient Boosting scores']
lst_feature_name = x.columns.values

feature_importance = pd.DataFrame()
for num in range(4):
  features = lst_mdl[num].feature_importances(X_train, y_train)
  feature_importance.insert(num,lst_mdl_name[num] ,features)
feature_importance.insert(0, 'feature', lst_feature_name)
```
